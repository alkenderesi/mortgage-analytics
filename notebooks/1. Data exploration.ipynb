## 1.7. Column descriptions

1. **AGE**: Age of customer (`int`)
2. **YEARS_WITH_BANK**: Number of years the customer has been with the bank (`int`)
3. **MARTIAL_STATUS**: Customer family state (`category`)
4. **EDUCATION**: Customer education level (`category`)
5. **EMPLOYMENT**: Customer emplyment status (`category`)
6. **GENDER**: Female or male (`category`)
7. **CUST_INCOME**: Customer monthly income in EUR (`float`)
8. **CURRENT_ADDRESS_DATE**: Date when customer moved to their current address (`date`)
9. **CURRENT_JOB_DATE**: Date when customer started their current job (`date`)
10. **CURRENT_WITH_BANK_DATE**: Date when customer opened their current bank account (`date`)
11. **CURRENT_BALANCE_EUR**: Current balance of customer in EUR (`float`)

12. **Mortgage_YN**: Y if customer has a mortgage, N if not (`category`)
13. **AGE_AT_ORIGINATION**: Age of customer when they took the mortgage (`int`)
